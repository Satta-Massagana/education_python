```python
Задание: написать программу, которая поможет пользователям определить, сколько денег они могут сэкономить на покупках
в разных магазинах.

Условия:
1) Пользователь должен ввести список товаров, которые он собирается купить, и цены на эти товары в разных магазинах.
2) Затем программа должна вывести список магазинов и общую стоимость покупок в каждом магазине, а также сообщить, в каком 
магазине пользователь может сэкономить больше всего денег.
3) Cписок товаров и магазинов определяется самостоятельно.
```

```python
import random

# Создаем список товаров с их названиями
products = [
    "Смартфон 'Технофон X'",
    "Ноутбук 'Элитбук Pro'",
    "Телевизор 'Мегавид 4K'",
    "Кофемашина 'Ароматно'",
    "Холодильник 'Фрешмейкер'"
]

# Вывод списка товаров
print("Товары в наличии:")
for product in products:
    print(product)
print()

# Запрос товаров и их количества у пользователя
user_shopping = {}
for product in products:
    quantity = int(input(f"Сколько {product} вы хотите приобрести: "))
    user_shopping[product] = quantity

# Создаем словарь для каждого магазина с разной стоимостью товаров
store1 = {product: random.randint(50, 300) for product in products}
store2 = {product: random.randint(50, 300) for product in products}
store3 = {product: random.randint(50, 300) for product in products}

# Создаем словарь, где ключами являются названия магазинов, а значениями - структуры данных для товаров
stores = {
    "Электроника Маркет": store1,
    "ТехноТрейд": store2,
    "Домашние Бытовые Товары": store3
}

# Создаем словарь для записи вывода
output_data = {}

# Записываем информацию о товарах и ценах в словарь
for store, products in stores.items():
    store_data = {}
    total_cost = 0
    for product, price in products.items():
        if product in user_shopping:
            total_cost += user_shopping[product] * price
        store_data[product] = price
    output_data[store] = {
        "Products": store_data,
        "Total Cost": total_cost
    }

# Выводим информацию о товарах и ценах в каждом магазине и общую стоимость покупок
print()
print("Информация о товарах и ценах в каждом магазине:")
print()
for store, data in output_data.items():
    print(f"{store}:")
    for product, price in data["Products"].items():
        print(f"{product}: {price} рублей")
    print(f"Общая стоимость: {data['Total Cost']} рублей")
    print()

# Определяем, в каком магазине пользователь может сэкономить больше всего денег
max_savings = min(output_data, key=lambda store: output_data[store]["Total Cost"])
print(f"Вы можете сэкономить больше всего денег, выбрав магазин '{max_savings}'.")
```
